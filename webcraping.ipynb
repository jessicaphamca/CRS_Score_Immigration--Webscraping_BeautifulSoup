```python
import requests
```

```python
page = requests.get("https://www.canada.ca/en/immigration-refugees-citizenship/corporate/mandate/policies-operational-instructions-agreements/ministerial-instructions/express-entry-rounds.html")
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(page.content, 'html.parser')
```

```python
soup.find_all('tr')
```

```python
col_headers_data = [item.get_text() for item in soup.find_all('tr')][0]
col_headers_data
```

```python
col_headers_data = [item.get_text() for item in soup.find_all('tr')][0]
col_headers = col_headers_data.split('\n')[1:-1]
col_headers
```

```python
row_data = [[item.get_text() for item in dat_h.find_all('td')] for dat_h in soup.find_all('tr')][1:-1]
row_data
```

```python
import pandas as pd
df = pd.DataFrame(row_data, columns=col_headers)
pd.set_option('display.max_rows', None)
df.head(100)
```
